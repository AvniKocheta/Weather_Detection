```python
df = pd.DataFrame(result)
df.reset_index(inplace=True)
# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'], format='mixed', dayfirst=True)

# Extract month from the datetime column
df['Month'] = df['Date'].dt.month
df['Month'] = df['Date'].dt.strftime('%B')

fig, axes = plt.subplots(nrows=6, ncols=1, figsize=(18, 20))

# Flatten the axes for easier iteration
axes = axes.flatten()

# Plot each feature on a separate subplot
features_to_plot = ['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Press_kPa', 'Weather']

for i, feature in enumerate(features_to_plot):
    sns.lineplot(data=df, x='Month', y=feature, ax=axes[i])
    axes[i].set_title(f'{feature} over Months')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
x = pd.get_dummies(df, columns = ['Weather'], dtype =int)
x.head()
```

```python
x.columns
```

```python
x.reset_index(inplace=True)
# Feature Engineering: Extract relevant features from the date
x['Year'] = x['Date'].dt.year
x['Month'] = x['Date'].dt.month
x['Day'] = x['Date'].dt.day
x['DayOfWeek'] = x['Date'].dt.dayofweek

# Drop the original 'Date' column
x.drop(columns=['Date'], inplace=True)

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(x[['Year', 'Month', 'Year', 'DayOfWeek','Press_kPa', 'Temp_C', 'Dew Point Temp_C', 'Rel Hum_%',
       'Wind Speed_km/h', 'Visibility_km']], x[['Weather_Clear',
       'Weather_Cloudy', 'Weather_Drizzle,Fog', 'Weather_Drizzle,Snow,Fog',
       'Weather_Fog', 'Weather_Haze', 'Weather_Mainly Clear',
       'Weather_Mostly Cloudy', 'Weather_Rain', 'Weather_Rain Showers',
       'Weather_Rain,Fog', 'Weather_Snow', 'Weather_Snow,Blowing Snow',
       'Weather_Snow,Fog', 'Weather_Snow,Haze']], test_size=0.3, random_state=42)

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = model.predict(X_test)
```

```python
# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y_test, Y_pred)
print(f'Mean Squared Error: {mse}')

# Calculate R-squared
r2 = model.score(X_test, Y_test)
print(f'R-squared: {r2}')

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(Y_test, Y_pred)
print(f'Mean Absolute Error: {mae}')
```

```python
new_data = [[2024, 3, 12, 5, 100, 20, 15, 70, 10, 10]]  # Example new data point
predicted_weather = model.predict(new_data)

print("Predicted weather probabilities:")
print(predicted_weather)
```

```python
from sklearn.metrics import accuracy_score

# Convert predicted probabilities to binary predictions
Y_pred_binary = (Y_pred > 0.37).astype(int)  

# Compute accuracy
accuracy = accuracy_score(Y_test, Y_pred_binary)
print(f'Accuracy: {accuracy}')
```
